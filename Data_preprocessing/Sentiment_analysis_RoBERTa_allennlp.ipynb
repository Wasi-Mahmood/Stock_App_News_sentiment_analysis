```python
import pandas as pd
df2 = pd.read_csv('Data_searching\Apple_dataset.csv')
```

```python
import spacy
import en_core_web_lg
nlp = en_core_web_lg.load()
```

```python
from text_processing import text_to_paragraph
```

```python
def text_normalization_for_sentiment_analysis(text_list, spacy_nlp_obj):
    sentence_list =[]
    for text in text_list:
        doc = spacy_nlp_obj(text)
        norm_text =[]
        for token in doc:
            if not token.is_punct and not token.is_space:
                norm_text.append(token.lemma_)
        sentence_list.append(' '.join(norm_text))
        
    return sentence_list
```

```python
import numpy as np


def compute_sentiment_RoBERTa(text_arr,min_text_len):
    positive_sentm =[]
    negative_sentm = [] 

    #text2 ="very good excellent apple stock grew by 50%"
    for index,text in enumerate(text_arr):
        if index != 0 and len(text)<min_text_len:
            break
        else:
            preds = predictor.predict(text)
            positive_sentm.append(round(preds['probs'][0],3))
            negative_sentm.append(round(preds['probs'][1],3))
        
        #print(f"p(positive)={preds['probs'][0]:.0%}\nn(negative)={preds['probs'][1]:.0%}")

    try:
        positive_sentm =np.mean(positive_sentm)
        negative_sentm = np.mean(negative_sentm)
    except:
        print("Can't compute mean")

    return positive_sentm, negative_sentm
```

```python
def add_sentm_values_to_dataframe(dataframe):
    #count =0
    for count in range(len(dataframe)):
        text = dataframe['Artical'][count]
        if pd.isnull(text):
            dataframe['pos_sentm'][count] =0
            dataframe['neg_sentm'][count] =0
        else:
            text_list = text_to_paragraph(text,1900,nlp)
            text_list2= text_normalization_for_sentiment_analysis(text_list,nlp)
            pos_sentm, neg_sentm = compute_sentiment_RoBERTa(text_list2,900)
            dataframe['pos_sentm'][count] = pos_sentm
            dataframe['neg_sentm'][count] = neg_sentm
        
        #if count ==4:
        #    break
        #count +=1
    return dataframe
```

```python
def text_to_paragraph(text,paragraph_len,spacy_nlp_obj):
    paragraph=""
    sentence_list =[]
    check_if_appened = False

    if len(text) <= paragraph_len:
        sentence_list.append(text)
        
    elif len(text)>= paragraph_len:
        
        doc = spacy_nlp_obj(text)
        
        for sents in doc.sents:
            if len(paragraph)<= paragraph_len:
                paragraph = paragraph + str(sents)
            if len(paragraph) >= paragraph_len:
                sentence_list.append(paragraph)
                paragraph = ""
                check_if_appened = True
                
        if check_if_appened is False:
            sentence_list.append(paragraph)
        
        if str(sentence_list[-1]) != paragraph:
            sentence_list.append(paragraph)
            
    return sentence_list
```
